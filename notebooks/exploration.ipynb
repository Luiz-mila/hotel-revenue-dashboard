```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv('hotel_reservations.csv')

print("=" * 60)
print("üìä DATASET OVERVIEW")
print("=" * 60)

print(f"\n Size: {df.shape[0]} rows x {df.shape[1]} columns")

# Available columns
print("\nüìã Available columns:")
for i, col in enumerate(df.columns, 1):
    print(f"   {i}. {col}")

# Preview first rows
print("\n First 5 rows:")
display(df.head())
```

```python
print("=" * 60)
print("üìã COLUMN INFORMATION")
print("=" * 60)

# Data type of each column
print("\nüîç Data types:")
print(df.dtypes)

# Check for missing values (NaN)
print("\n‚ùì Missing values per column:")
print(df.isnull().sum())

# Basic statistics for numerical columns
print("\nüìä Numerical statistics:")
display(df.describe())
```

```python
# Check unique values for key categorical columns
print("=" * 60)
print("üîç UNIQUE VALUES OF KEY COLUMNS")
print("=" * 60)

# Meal plan type
print("\nüçΩÔ∏è Meal plan types:")
print(df['type_of_meal_plan'].value_counts())

# Room type
print("\nüõèÔ∏è Room types:")
print(df['room_type_reserved'].value_counts())

# Market segment
print("\nüìä Market segments:")
print(df['market_segment_type'].value_counts())

# Booking status
print("\n‚úÖ Booking status:")
print(df['booking_status'].value_counts())
```

```python
# ================================================
# DATA CLEANING & FEATURE ENGINEERING
# ================================================

print("=" * 60)
print("STARTING DATA CLEANING")
print("=" * 60)

print(f"\nüìä Original dataset: {len(df)} rows")

# 1. Remove invalid bookings (0 adults)

df_clean = df[df['no_of_adults'] > 0].copy()
print(f"\n‚úÖ After removing 0 adults: {len(df_clean)} rows")
print(f"Removed: {len(df) - len(df_clean)} rows")

# 2. Remove bookings with 0 nights

df_clean = df_clean[
    (df_clean['no_of_weekend_nights'] > 0) |
    (df_clean['no_of_week_nights'] > 0)
].copy()

print(f"\nAfter removing 0 nights: {len(df_clean)} rows")
print(f"Removed: {len(df) - len(df_clean)} rows")

# 3. Create useful columns

# Total nights stayed
df_clean['total_nights'] = (
    df_clean['no_of_weekend_nights'] +
    df_clean['no_of_week_nights']
)

# Total revenue (price per room * nights)
df_clean['total_revenue'] = (
    df_clean['avg_price_per_room'] *
    df_clean['total_nights']
)

# Total guests
df_clean['total_guest'] = (
    df_clean['no_of_adults'] +
    df_clean['no_of_children']
)

# Revenue per guest
df_clean['revenue_per_guest'] = (
    df_clean['total_revenue'] /
    df_clean['total_guest']
)


print(f"\n‚úÖ Created 4 new columns:")
print(f"   - total_nights")
print(f"   - total_revenue")
print(f"   - total_guests")
print(f"   - revenue_per_guest")

# 4. Check final dataset

print(f"\nüìä Final dataset: {len(df_clean)} rows x {len(df_clean.columns)} columns")

# Show sample with new columns
print(f"\n Sample with new columns:")

sample_cols = [
    'Booking_ID', 
    'no_of_adults', 
    'total_nights', 
    'avg_price_per_room',
    'total_revenue',
    'booking_status',
    'total_guest',
    'revenue_per_guest'
]

display(df_clean[sample_cols].head(3))
```

```python
# ================================================
# CALCULATE KEY PERFORMANCE INDICATORS (KPIs)
# ================================================

"""
HOTEL REVENUE MANAGEMENT KPIs - INDUSTRY BENCHMARKS

ADR (Average Daily Rate):
    - Budget Hotels: $50-80
    - Mid-Range Hotels: $80-150
    - Upscale Hotels: $150-300
    - Luxury Hotels: $300+

Cancellation Rate:
    - Excellent: < 15%
    - Good: 15-25%
    - Average: 25-35%
    - Poor: > 35% (requires investigation)

Average Length of Stay (ALOS):
    - Business Hotels: 1-2 nights
    - City Hotels: 2-3 nights
    - Resort Hotels: 4-7 nights
    - Extended Stay: 7+ nights

Repeat Guest Rate:
    - Excellent: > 30%
    - Good: 20-30%
    - Average: 10-20%
    - Poor: < 10%

Lead Time (Booking Window):
    - Last Minute: < 7 days
    - Short Notice: 7-30 days
    - Standard: 30-90 days
    - Early Booking: > 90 days

RevPAR (Revenue Per Available Room):
    - Budget: $30-60
    - Mid-Range: $60-120
    - Upscale: $120-250
    - Luxury: $250+
    
Note:  
   RevPAR (Revenue Per Available Room) cannot be calculated
   because this dataset does not include:
   - Total available rooms in the hotel
   - Actual occupancy rate
   
   Alternative metric used: Revenue per Night Sold
   This measures efficiency of sold inventory rather than
   total available inventory.
"""

print("=" * 60)
print("üìä KEY PERFORMANCE INDICATORS (KPIs)")
print("=" * 60)

# 1. Overall Revenue Metrics

total_bookings = len(df_clean)
confirmed_bookings = len(df_clean[df_clean['booking_status'] == 'Not_Canceled'])
canceled_bookings = len(df_clean[df['booking_status'] == 'Canceled'])

""" Only use confirmed bookings for revenue """
df_confirmed = df_clean[df_clean['booking_status'] == 'Not_Canceled'].copy()

total_revenue = df_confirmed['total_revenue'].sum()
adr = df_confirmed['avg_price_per_room'].mean()
total_room_nights = df_confirmed['total_nights'].sum()

print(f"\n --- REVENUE METRICS:")
print(f"      Total Revenue: {total_revenue:,.2f}")
print(f"      Average Daily Rate (ADR): ${adr:,.2f}")
print(f"      Total Room Nights Sold: {total_room_nights:,}")

# ADR Benchmark classification
if adr < 80:
    adr_category = "Budget"
elif adr < 150:
    adr_category = "Mid-Range"
elif adr < 300:
    adr_category = "Upscale"
else:
    adr_category = "Luxury"
print(f"      ADR Category: {adr_category}")

# 2. Booking Performance

cancellation_rate = (canceled_bookings / total_bookings) * 100

print(f"\n --- BOOKING PERFORMANCE:")
print(f"      Total Bookings: {total_bookings}")
print(f"      Confirmed: {confirmed_bookings}")
print(f"      Canceled: {canceled_bookings}")
print(f"      Cancelation Rate: {cancellation_rate:.1f}%")

# Cancellation rate benchmark
if cancellation_rate < 15:
    cancel_rating = "Excellent"
elif cancellation_rate < 25:
    cancel_rating = "Good"
elif cancellation_rate < 35:
    cancel_rating = "Average ‚ö†Ô∏è"
else:
    cancel_rating = "Poor  (Requires Investigation)"
print(f"      Performance: {cancel_rating}")

# 3. Guest Metrics

avg_length_of_stay = df_confirmed['total_nights'].mean()
avg_guests_per_booking = df_confirmed['total_guest'].mean()
repeat_guest_rate = (df_confirmed['repeated_guest'].sum() / len(df_confirmed)) * 100

print(f"\n --- GUEST METRICS:")
print(f"      Average Length of Stay: {avg_length_of_stay:.1f}")
print(f"      Average Guests per Booking: {avg_guests_per_booking:.1f}")
print(f"      Repeat Guest Rate: {repeat_guest_rate:.1f}%")

# ALOS hotel type classification
if avg_length_of_stay < 2:
    hotel_type = "Business Hotel"
elif avg_length_of_stay < 4:
    hotel_type = "City Hotel"
elif avg_length_of_stay < 7:
    hotel_type = "Resort Hotel"
else:
    hotel_type = "Extended Stay"
print(f"      Hotel Type (by ALOS): {hotel_type}")

# Repeat guest benchmark
if repeat_guest_rate > 30:
    repeat_rating = "Excellent"
elif repeat_guest_rate > 20:
    repeat_rating = "Good"
elif repeat_guest_rate > 10:
    repeat_rating = "Average ‚ö†Ô∏è"
else:
    repeat_rating = "Poor ‚ùå"
print(f"      Loyalty Performance: {repeat_rating}")

# 4. Advanced Metrics

avg_lead_time = df_confirmed['lead_time'].mean()
revenue_per_available_night = total_revenue / total_room_nights

print(f"\n --- ADVANCED METRICS:")
print(f"      Average Lead Time: {avg_lead_time:.0f} days")
print(f"      Revenue per Night Sold: ${revenue_per_available_night:.2f}")

# Lead time classification
if avg_lead_time < 7:
    booking_pattern = "Last Minute"
elif avg_lead_time < 30:
    booking_pattern = "Short Notice"
elif avg_lead_time < 90:
    booking_pattern = "Standard Planning"
else:
    booking_pattern = "Early Booking"
print(f"      Booking Pattern: {booking_pattern}")
```

```python
# ================================================
# EXPORT CLEANED DATA FOR SQL IMPORT
# ================================================

print("=" * 60)
print("üíæ EXPORTING CLEANED DATA")
print("=" * 60)

# Save cleaned dataset
output_path = '../data/hotel_reservations_clean.csv'
df_clean.to_csv(output_path, index=False)

print(f"\n‚úÖ Cleaned data saved to: {output_path}")
print(f"   Total rows: {len(df_clean):,}")
print(f"   Total columns: {len(df_clean.columns)}")

# Show columns that will be imported to SQL
print(f"\nüìã Columns to import:")
for i, col in enumerate(df_clean.columns, 1):
    dtype = df_clean[col].dtype
    print(f"   {i:2d}. {col:30s} ({dtype})")
```
